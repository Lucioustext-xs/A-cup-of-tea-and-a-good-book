```python
def getQuote():
  quotes = []

  URL = 'https://justlittlethings.net/list'
  webpage = requests.get(URL) #Make a request to the web 

  soup = BeautifulSoup(webpage.text, 'html.parser') #Parse the text from it
  quoteText = soup.find_all('div', attrs={'id': 'box'})
  
  for i in quoteText:
    for j in range(3,2000):
  #quote = i.text.strip().split('\xa0')[1]
      quote = (i.text.strip().split('.')[j])
      quotes.append(quote)
  #print(i.text.strip().split('\xa0')[1])

  quotes = [re.sub('[^a-zA-Z0-9]+', ' ', _) for _ in quotes]
  quotes = [item for item in quotes if not item.isdigit()]
  jsonString = json.dumps(quotes)

  return jsonString



```
